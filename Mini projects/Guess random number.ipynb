# Guess the number

Guess the number that is randomly generated by the computer, which keeps changing on each try

```python
import random

print('''
Welcome to the number guessing game! 
Please enter a number between 1 and 100. 
You have 5 chance to guess the number. Good luck!
''')

num = random.randrange(100);
guess = 0;
count = 0;
limit = 5;

while guess!= num and count < limit:
 guess = int(input("Guess the number: "))
 count+= 1
 if guess == num:
     print("You won!")
     break
 elif count == limit:
    print("Out of guesses! The correct number is " + str(num) + ". Better luck next time!")
    break
 elif guess > num:
     print("Try again! Your guess is higher than the number!")
 elif guess < num:
     print("Try again! Your guess is lesser than the number!")
```
