# Data Cleaning

```python
import pandas as pd

df = pd.read_excel(r"C:\Users\thahi\Data Analysis\Project\Pandas\Customer Call List.xlsx")
df.head()
```

```python
df.shape
```

Drop the duplicate the rows

```python
df.drop_duplicates(inplace = True)
```

```python
df.shape
```

Drop the unwanted columns

```python
df = df.drop(columns= 'Not_Useful_Column')
```

### Standardize the data:

Remove the unwanted characters from the column 'Last_Name'


```python
df['Last_Name'] = df['Last_Name'].str.strip('./_')
df
```

Format 'Phone_Number' column:
- Convert the dtype to string
- Remove unwanted characters 
- Convert them into the same format


```python
df['Phone_Number'] = df['Phone_Number'].astype(str)
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','', regex=True)
df.head()
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df.head()
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--','').str.replace('Na--','')
df
```

Split 'Address' column using commas seperator

```python
df[['Street', 'State', 'Zipcode']]=df['Address'].str.split(',', n=2, expand=True)
df.head()
```

Format the columns 'Paying Customer' and 'Do_Not_Contact' to have the same options	

```python
df['Paying Customer'] = df['Paying Customer'].replace({'No': 'N', 'Yes': 'Y'})
df.head()
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace({'No': 'N', 'Yes': 'Y'})
df.head()
```

```python
df = df.replace('N/a','')
df
```

Work with Null values

```python
df.isnull().sum()
```

```python
df = df.fillna('')
df
```

**In Case the client wants the details of only the customers with the valid phone numbers,** let us remove the rows with empty phone numbers

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('', 'N')
df
```

```python
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
       df.drop(x, inplace = True)

df
```

Reset the index column after dropping the unwanted rows

```python
df.reset_index(drop=True)
```
